```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Car data saved-1.xlsx")
df.head()
```

```python
variables=['Min Price (Lakh)','Max Price (Lakh)','Range (kmpl)',	'CC',
'Seats','Variants','Ex-Showroom Price','RTO','Insurance',
'Other','Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i,var in enumerate(variables, start=1):
    plt.subplot(4,3,i)
    sns.scatterplot(x=df[var],y=df['Max Price (Lakh)'],alpha=0.7)
    plt.title(f'scatter plot of{var}vs Onroad price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i,var in enumerate(variables, start=1):
    plt.subplot(3,4,i)
    sns.histplot(df[var],bins=20, kde =True, color='purple')
    plt.title(f'Histogram of{var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#create a boxplot of range by car type
sns.boxplot(data=df,x='Type',y='Range (kmpl)',hue = 'Type', palette =['purple','pink'])
plt.title('Boxplot of Range by Car Type')
plt.show()
```

```python
selected_columns=['Min Price (Lakh)',	'Max Price (Lakh)',	'Range (kmpl)',	'CC', 'Seats',
'Variants',	'Ex-Showroom Price',	'RTO',	'Insurance',	'Other']
df_selected = df[selected_columns]
df_selected.head()
```

```python
corr = df_selected.corr()
corr
```

```python
sns.heatmap(corr, cmap='BuPu')
plt.title('Heatmap of correlation Matrix')
plt.show()
```

```python
#violin plot of Max Price  by Car Type
sns.violinplot(data=df,x='Type', y='Max Price (Lakh)',hue = 'Type', palette =['purple','pink'])
plt.title('Violin Plot of Maximum Price by Car Type')
plt.show()
```
