```python
import pandas as pd
import matplotlib.pyplot as plt#loading csv file to jupyter notebook
import seaborn as sns
```

```python
data=pd.read_csv("iris.csv")
```

```python
data.head()
```

```python
df=data.copy()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
###FEATURE ENGINEERING/Processing of data
```

```python
 #map species to numerical values
df['species']=df['species'].astype('category').cat.codes
df.head()
```

```python
sns.pairplot(df)
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('visualize correlation')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
#features and target variables
x=df.drop('species',axis=1)
y=df['species']
#split the dataset into cleaning and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#import the logisticRegression class from sklearng
from sklearn.linear_model import LogisticRegression
#initialize the LogisticRegression model
logistic_model=LogisticRegression()
#Fit the model to the training data(x_train and y_train)
logistic_model.fit(x_train,y_train)
```

```python
y_pred_logistic = logistic_model.predict(x_test)#predict the target variable for the target data(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Logistic Regression Results:")
print("Accuracy:",accuracy_score(y_test,y_pred_logistic))
```

```python


plt.scatter(y_test,y_pred_logistic)
plt.xlabel('y_test')
plt.ylabel('y_pred_logistic')
plt.show()
```
