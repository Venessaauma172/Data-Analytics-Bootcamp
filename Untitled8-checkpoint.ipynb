```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Datafiles/datawhoo.xlsx")
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date']) #convert to datetime format

 #plotting the data
plt.plot(df['Date'], df['Sales'],color = "yellow")
plt.title('Sales Over Time')
plt.ylabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()       
```

```python

 
plt.bar(df['Product'], df['Sales'],color = "purple")
plt.title('Sales Per Product')
plt.ylabel('product')
plt.ylabel('Sales')
plt.show()       
```

```python
data = pd.read_excel("Car data saved-1.xlsx")
data.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data["Max Price (Lakh)"],data['Insurance'],color="purple")
plt.title('Max Price(lakh) vs insurance')
plt.xlabel("Max Price (Lakh)")
plt.ylabel('Insurance')
plt.grid(True)
plt.show()       
```

```python
df = pd.read_excel("Datafiles/datawhoo.xlsx")
df.head()
```

```python
sales_data=df.groupby('Category')['Sales'].sum()
sales_data
```

```python
plt.figure(figsize=(8,8))#Create the pie chart
plt.pie(sales_data,labels=sales_data.index,autopct='%1.1f%%')

plt.title('proportion of Sales by Category')
plt.show()
```

```python
plt.hist(df['Sales'],bins=5, color='blue',edgecolor='red')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()       
```

```python
plt.fill_between(df['Date'],df['Sales'],color='blue',alpha=0.3)
plt.plot (df['Date'],df['Sales'],color='lightgreen')
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.plot(df['Country'], df['Country'],color = "yellow")
plt.title('Population')
plt.xlabel('Country')
plt.ylabel('County')
plt.xticks(rotation=45)
plt.show()       
```
